## Clustering
- Clustering(군집화)는 유사한 속성을 가지는 데이터를 특정 기준에 따라 묶는 것
- 좋은 군집화는 동일한 군집 속에 있는 데이터는 서로 유사하고, 다른 군집 속에 있는 데이터는 서로 다른 것
- 군집화는 크게 계층적 군집화(Hierarchical Clustering), 분할적 군집화(Partitional Clustering) 두 가지로 나뉜다

---

## 계층적 군집 (Hierarchical Clustering)



이미지 출처: https://tyami.github.io/machine%20learning/hierarchical-clustering/#types-of-hierarchical-clustering

- 합병형 - 새로운 개체를 본 데이터에 하나씩 추가하는 방식
-  AGNES (AGglomerative NESting, __Bottom-up__ approach)이라고도 불림

> 단일 (최단) 연결법
> > - 데이터를 기준으로 최단 거리, 즉 가장 가까운 데이터가 있는 군집에 포함 시킴

> 완전 (최장) 연결법
> > - 데이터를 기준으로 최장 거리, 즉 가장 먼 데이터가 있는 군집에 포함 시킴

> 평균 연결법
> > - 각 군집에 속한 데이터들과의 거리 평균을 구하고 평균값이 가장 작은 군집에 포함 시킴

> 와드 연결법
> > - 새로운 개체를 포함 시킴으로써 늘어나는 오차가 가장 작은 군집에 포함 시킴

> 중심 연결법
> > - 새로운 개체와 각 군집의 중심까지 거리가 짧은 군집에 포함 시킴 

- 분리형 - 전체 데이터를 하나의 군집으로 시작하여 개체들을 하나씩 분리해나가는 방식
- DIvise ANAlysis, __Top-down__ approach
> 다이애나(DIANA) 방법

## 분할적 군집 (Partitional Clustering)
- 프로토타입 기반 (기초, 표준 기반)

### K-centroid 
- K-means, K-median, K-medoid를 모두 묶어서 K-centroid(K중심군집)이라고 부름
- 연속형 데이터의 경우  평균(mean)이나  중앙값(median)을 그 군집의 프로토타입으로 하며, 
- 이산형 데이터의 경우  최빈값(mode)이나  메도이드(medoid)를 프로토타입으로 정함

|  데이터의 종류  | 사용 방법 |
|:---------------:|:----------:|
|    연속형      | K-means, K-median|
|    이산형      | K-medoid      |

### K-means
- 데이터를 특정 기준에 따라 K개의 Cluster로 묶는 알고리즘
- 특정 기준으로 데이터들을 비교. sklearn은 euclidean distance가 default
- cosine similarity를 기준으로 한 SphericalKMeans도 존재

> K-means 학습 방식?
> > - 1. cluster centers를 랜덤하게 생성
> > - 2. 각각의 개체마다 어떤 cluster와 가장 가까운지 계산 및 할당 
> > - 3. 새로운 label을 이용해서, cluster_centers 새로 계산 
> > - 4. 클러스터에 변화가 없으면 종료

> __K-means__ vs __KNN__

|  방법 | 공통점|차이점|
|-----|--------|---|
|K-means|K개의 점을 지정하여 거리를 기반으로 구현되는 거리 기반 분석 알고리즘 |비지도학습|
|  KNN  |K개의 점을 지정하여 거리를 기반으로 구현되는 거리 기반 분석 알고리즘 |지도학습|

> 단점
> > - 초기의 중심점을 랜덤하게 설정하기 때문에 초기 중심값에 따라 clustering 결과가 많이 달라질 수 있음
> > - 즉, 돌릴 때마다 결과가 달라짐
> > - Local optimum에 빠질 수 있음
> > - 이상치에 따라 결과가 많이 달라짐. 즉, robust하지 않음

### K-median 
- K-median은 K-Means와 관련되어 있긴 하지만 mean을 구할 때 그룹의 median vector를 사용해 mean을 구함

### K-medoid
- 초기 중심 값에 민감한 반응을 보이고, noise와 이상치에 민감했던 k-means의 단점을 보완 
- 군집의 중심을 구하기 위해서 데이터의 평균대신 중간점(medoids)를 사용 
- 그렇기에 데이터의 이상치에 대해서 더 강건함. 즉, robust함

### 퍼지 군집 (Fuzzy clustering)
- Fuzzy: 애매함, 모호함의 뜻. 즉, 어떤 군집에 속할지 애매한 상황에 좋음
- 각 개체가 한 개 이상의 군집에 속할 수 있도록 군집을 형성하는 것
- 각 데이터가 군집에 속할 확률에 대해 정보를 전달해주는 것을 목적으로 하는 군집화
- ex) 하나의 개체가 A에 속할 확률이 51, B에 속할 확률이 49인 경우 이 정보를 전달해줌

> Hard Clustering & Soft Clustering

|   Hard Clustering   |   Soft Clustering   |
|        ---          |         ---         |
|각 개체가 반드시 하나의 군집에 속함|한개 이상의 군집에 속할 수 있음|
|해당 군집에 속하면 1, 아니면 0으로 할당|해당 군집에 속할 확률값을 알려줌|
